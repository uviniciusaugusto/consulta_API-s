```python
import json
import urllib
```

```python
usuario = input('Usuario: ')
dadosPessoais = {}
dadosRepositorios = {}
nomeSeguidores = []
loginSeguidores = []
```

```python
def get_dados_pessoais(nome):
    if nome not in dadosPessoais:
        url =  'https://api.github.com/users/'+nome
        retorno = urllib.request.urlopen(url).read()
        dadosPessoais[usuario] = json.loads(retorno)
```

```python
def get_repositorios(nome):
    if nome not in dadosRepositorios:
        url =  'https://api.github.com/users/'+nome+'/repos?type=owner'
        retorno = urllib.request.urlopen(url).read()
        dadosRepositorios[usuario] = json.loads(retorno)
```

```python
def get_seguidores(nome):
    url =  'https://api.github.com/users/'+nome+'/followers'
    retorno = urllib.request.urlopen(url).read()
    seguidores = json.loads(retorno)
    for seguidor in seguidores:
        get_dados_pessoais(seguidor['login'])
        print("\n" + str(dadosPessoais[nome]['name']))
        get_repositorios(seguidor['login'])
        for repositorio in dadosRepositorios[nome]:
            print("\t" + str(repositorio['name']))
        
```

```python
get_dados_pessoais(usuario)
dadosPessoais[usuario]

print("Usu√°rio da consulta: " + usuario)
print("\n" + str(dadosPessoais[usuario] ['followers']) + " seguidores")
get_seguidores(usuario)
```
