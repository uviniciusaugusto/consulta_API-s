```python
import json
import urllib
```

```python
API = 'https://viacep.com.br/ws/{0}/json'
cep = input('Fornceça o CEP: ')
dados = {}
```

```python
if cep not in dados:
    retorno = urllib.request.urlopen(API.format(cep)).read()
    dados[cep] = json.loads(retorno)
```

```python
try:
    print ('Bairro: ', dados[cep]['bairro'])
    print ('DDD: ', dados[cep]['ddd'])
    print ('Estado: ', dados[cep]['uf'])
    print ('Cidade: ', dados[cep]['localidade'])
except:
     print ('CEP inválido ou erro na API')
```
